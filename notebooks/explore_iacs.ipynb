# Exploring iacs

This notebook demonstrates the core iacs functionality for working with ECS data.

```python
from pathlib import Path

from iacs.io_system import IOSystem
from iacs.registry import Registry
```

## Reading Entity-Centered Data

The `IOSystem` reads entity-centered data (the human-friendly format) into a DataFrame.

```python
# Create an IOSystem instance
io = IOSystem()

# Read from a dict
data = {
    "my_task": [
        {"description": "A task I need to complete."},
        "task",
    ],
    "my_infrastructure": [
        {"description": "Infrastructure to complete the task."},
        {"implements": "my_task"},
    ],
}

entity_centered = io.read_entity_centered(data)
entity_centered
```

## Reading from Files

You can also read entity-centered data directly from YAML files.

```python
# Read the minimal example
minimal_df = io.read_entity_centered_file(Path("../examples/minimal.yaml"))
minimal_df
```

```python
# Read the hierarchical example (minimal2)
minimal2_df = io.read_entity_centered_file(Path("../examples/minimal2.yaml"))
minimal2_df
```

```python
# Read a network example
net_ab_df = io.read_entity_centered_file(Path("../examples/networks/net_AB.yaml"))
net_ab_df
```

## Converting to Component-Centered Format (Registry)

The `Registry` stores data in component-centered format, with one table per component type.

```python
# Convert entity-centered data to a Registry
registry = Registry.from_entity_centered(entity_centered)

# See what component types are in the registry
registry.component_types
```

```python
# View the description component table
registry.view("description")
```

```python
# View the task component table (tag components have no value column)
registry.view("task")
```

```python
# View the implements component table
registry.view("implements")
```

## Working with the Components YAML

Load and explore the main components.yaml file that defines iacs itself.

```python
# Read components.yaml
components_df = io.read_entity_centered_file(Path("../components/components.yaml"))
print(f"Total components: {len(components_df)}")
print(f"Unique entities: {components_df['entity_id'].nunique()}")
print(f"Component types: {components_df['component_type'].unique().tolist()}")
```

```python
# Convert to registry
components_registry = Registry.from_entity_centered(components_df)
components_registry.component_types
```

```python
# View all requirements
components_registry.view("requirement")
```

```python
# View all todos
components_registry.view("todo")
```
